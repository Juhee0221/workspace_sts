```python
import pymysql
import pandas as pd
```

```python
team_df = pd.read_csv('TeamAPI.csv', encoding='cp949')
```

```python
from datetime import datetime
# # print(datetime.today().strftime("%Y-%m-%d"))
# print(team_df['데이터 시간'].str.slice(start=0, stop=10) == datetime.today().strftime("%Y-%m-%d"))
my_filter = team_df['데이터 시간'].str.slice(start=0, stop=10) == datetime.today().strftime("%Y-%m-%d")
new_df = team_df[my_filter]
```

```python
staion_df = pd.read_excel('Station.xlsx')
staion_df
```

```python
my_filter_2 = staion_df['시리얼번호'].isin(new_df['시리얼번호'])
new_station_df = staion_df[my_filter_2]

new_station_df.isnull().sum()

new_station_df.index = range(len(new_station_df))

new_station_df
```

```python
conn = pymysql.connect(host='localhost', user='studyroom', password='studyroom', db='studyroom', charset='utf8')
cursor = conn.cursor(pymysql.cursors.DictCursor)
```

```python
# Map 만들기

for i in range(len(new_station_df)):
    mapData = f"INSERT INTO MAP (SERIAL_NO, STATION_NAME, LAT, LON) VALUES ('{new_station_df['시리얼번호'][i]}', '{new_station_df['스테이션명'][i]}', {new_station_df['위도'][i]}, {new_station_df['경도'][i]})"
    # 실행
    cursor.execute(mapData)

# commit 필수
conn.commit()
```

```python
conn.close()
```
