```python
import pymysql
import pandas as pd
```

```python
team_df = pd.read_csv('TeamAPI.csv', encoding='cp949')
team_df
```

```python
idxList = []

for idx, each in enumerate(team_df['미세먼지(/)']):
    print(idx, each)
    if not each.isdigit():
        idxList.append(idx)

idxList

team_df.drop(index=idxList, inplace=True)
```

```python
team_df.isnull().sum()
```

```python
team_df.index = range(len(team_df))
```

```python
team_df
```

```python
## 소음 등급 컬럼 추가

team_df.loc[(pd.to_numeric(team_df['소음(dB)']) >= 0) & (pd.to_numeric(team_df['소음(dB)']) <= 60), '소음 등급'] = '좋음'
team_df.loc[(pd.to_numeric(team_df['소음(dB)']) > 60) & (pd.to_numeric(team_df['소음(dB)']) <= 80), '소음 등급'] = '보통'
team_df.loc[(pd.to_numeric(team_df['소음(dB)']) > 80), '소음 등급'] = '나쁨'
team_df
```

연결 코드

```python
conn = pymysql.connect(host='localhost', user='studyroom', password='studyroom', db='studyroom', charset='utf8')
cursor = conn.cursor(pymysql.cursors.DictCursor)
```

시리얼번호 따로 저장

```python
SERIAL_NO = team_df['시리얼번호']
```

전체 평균 값 구하기

```python
# 평균값 모아놓을 리스트
avgList = []
```

```python
# (1) 미세먼지 전체 평균 구하기
fineDust = sum(pd.to_numeric(team_df['미세먼지(/)'])) / len(pd.to_numeric(team_df['미세먼지(/)']))
fineDustAvg = f"{fineDust:.2f}"
avgList.append(fineDustAvg)
```

```python
# (2) 초미세먼지 전체 평균 구하기
ultraFineDust = sum(pd.to_numeric(team_df['초미세먼지(/)'])) / len(pd.to_numeric(team_df['초미세먼지(/)']))
ultraFineDustAvg = f"{ultraFineDust:.2f}"
avgList.append(ultraFineDustAvg)
```

```python
# (3) 이산화탄소 전체 평균 구하기
team_df['이산화탄소(ppm)'].fillna(0, inplace=True)
co2 = sum(pd.to_numeric(team_df['이산화탄소(ppm)'])) / len(pd.to_numeric(team_df['이산화탄소(ppm)']))
co2Avg = f"{co2:.2f}"
avgList.append(co2Avg)
```

```python
# (4) 휘발성유기화합물 전체 평균 구하기
team_df['휘발성유기화합물(ppb)'].fillna(0, inplace=True)
voc = sum(pd.to_numeric(team_df['휘발성유기화합물(ppb)'])) / len(pd.to_numeric(team_df['휘발성유기화합물(ppb)']))
vocAvg = f"{voc:.2f}"
avgList.append(vocAvg)
```

```python
# (5) 온도 전체 평균 구하기
temperature = sum(pd.to_numeric(team_df['온도(C)'])) / len(pd.to_numeric(team_df['온도(C)']))
temperatureAvg = f"{temperature:.2f}"
avgList.append(temperatureAvg)
```

```python
# (6) 습도 전체 평균 구하기
humidity = sum(pd.to_numeric(team_df['습도(%)'])) / len(pd.to_numeric(team_df['습도(%)']))
humidityAvg = f"{humidity:.2f}"
avgList.append(humidityAvg)
```

```python
# (7) 소음 전체 평균 구하기
noise = sum(pd.to_numeric(team_df['소음(dB)'])) / len(pd.to_numeric(team_df['소음(dB)']))
noiseAvg = f"{noise:.2f}"
avgList.append(noiseAvg)
```

```python
# (8) 통합실내지수 전체 평균 구하기
cici = sum(pd.to_numeric(team_df['통합실내지수'])) / len(pd.to_numeric(team_df['통합실내지수']))
ciciAvg = f"{cici:.2f}"
avgList.append(ciciAvg)
```

실질적 insert 문법

```python
# DECIBEL_INFO 데이터 추가
DECIBEL = pd.to_numeric(team_df['소음(dB)'])
DECIBEL_GRADE = team_df['소음 등급']

for i in range(len(team_df)):
     decibelData = f"INSERT INTO DECIBEL_INFO (DECIBEL, DECIBEL_GRADE, SERIAL_NO) VALUES ({DECIBEL[i]}, '{DECIBEL_GRADE[i]}', '{SERIAL_NO[i]}')"
     #execute 실행
     cursor.execute(decibelData)

# commit 필수
conn.commit()
```

```python
# AVERAGE 데이터 추가
for i in range(0,8):
    avgData = f"INSERT INTO AVERAGE (AVERAGE_RESULT, AVERAGE_NAME) VALUES ({avgList[i]},'{team_df.columns[i+2]}')"
    #execute 실행
    cursor.execute(avgData)

# commit 필수
conn.commit()
```
